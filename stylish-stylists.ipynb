```python
# Load Data, which we got as an export from https://osm.li/Xt
json = 'export.geojson'
# We should proabably look into reading this as geoJSON, then we can also plot it on a map.
# But for now we just want to get the names out of the file...
```

```python
# Get all names from the exported geoJSON file
Names = []
file = open(json, 'r')
for line in file:
    if 'name' in line:
        Names.append(line.split(':')[1].strip().lstrip('"').rstrip('",'))
file.close()
```

```python
# Remove duplicates
# Based on https://stackoverflow.com/a/27305828
Names = list(sorted(set(Names)))
```

```python
# What did we get?
print('We have %s unique hairdresser names in our list.' % len(Names))
```

```python
# What are we looking for?
SillyStrings = ['haar', 'hair', 'schnitt']
```

```python
# Print dem silly nems
for n in Names:
    # Only search in lowercase names, but find all of the names in the list above
    # Based on https://stackoverflow.com/a/3389611
    if any(name in n.lower() for name in SillyStrings):
        print(n)
```

```python
# Print *all* names
for c, n in enumerate(Names):
    print('%04s/%s: %s' % (c+1, len(Names), n))
```
